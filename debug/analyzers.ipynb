```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head(5)
```

```python
df.dropna(subset=["SERIAL_NUMBER"],inplace=True)
```

```python
yt = df.drop(columns=["TECHNOLOGY","NAME","TYPE","PRESENT","CYCLE_COUNT","MODEL_NAME","MANUFACTURER","SERIAL_NUMBER","VOLTAGE_MIN_DESIGN","VOLTAGE_NOW","CURRENT_NOW","CHARGE_FULL_DESIGN","CHARGE_FULL","CHARGE_NOW"])
```

```python
yt.head(5)
```

```python
yt["DATE"] = pd.to_datetime(yt["DATE"],format='%m/%d/%y')
```

```python
curr_date=yt["DATE"][0]
curr_date=pd.to_datetime(curr_date)
```

```python
th_bef = curr_date - pd.DateOffset(days=3)
th_bef
```

```python
filtered_yt = yt.loc[(yt["DATE"]>=th_bef) & (yt["DATE"]<=curr_date)]
```

```python
filtered_yt
```

```python
filtered_yt = filtered_yt.reset_index(drop=True)
```

```python
filtered_yt["DATE"][0]
```

```python
filtered_yt.to_csv("recent_3_days.csv",index=False)
```

```python
def conv_html(ip,output):
    df = pd.read_csv(ip)
    html_rows = []
    for i,rows in df.iterrows():
        if i==0:
            html_rows.append("<tr>")
            for h in df.columns:
                html_rows.append(f"<th>{h}</th>")
            html_rows.append("</tr>")

        status_class = 'recent-usage-status' if rows["STATUS"] == "Active" or rows["STATUS"] == "Suspended" else ''
        html_row = f"<tr><td>{rows.DATE}</td><td>{rows.TIME}</td><td class={status_class}>{rows.STATUS}</td><td>{rows.SOURCE}</td><td>{rows.CAPACITY}</td><td>{rows.CAPACITY_LEVEL}</td></tr>"
        html_rows.append(html_row)
    html_table = '<table>' + '\n'.join(html_rows) + '</table>'

    with open(output,'w') as f:
        f.write(html_table)                
```

```python
conv_html("recent_3_days.csv","a.html")
```
